#Theory Questions

1.  What is hypothesis testing in statistics?
- **Hypothesis testing** in statistics is a formal process used to make decisions or inferences about population parameters based on sample data. It helps determine whether there is enough statistical evidence to support a particular belief or hypothesis about a population.

---

###  **Key Concepts**

1. **Null Hypothesis (H‚ÇÄ)**

   * This is the default assumption or status quo.
   * It states that **there is no effect or no difference**.
   * Example: H‚ÇÄ: Œº = 50 (The population mean is 50)

2. **Alternative Hypothesis (H‚ÇÅ or Ha)**

   * This is what you aim to prove.
   * It states that **there is an effect or a difference**.
   * Example: H‚ÇÅ: Œº ‚â† 50

---

###  **Steps in Hypothesis Testing**

1. **State the Hypotheses**

   * Define H‚ÇÄ and H‚ÇÅ clearly.

2. **Choose a Significance Level (Œ±)**

   * Common choices: 0.05, 0.01, or 0.10
   * This represents the probability of rejecting H‚ÇÄ when it is actually true (Type I error).

3. **Select and Compute the Test Statistic**

   * Examples: Z-test, t-test, chi-square, ANOVA, etc.
   * The test statistic tells how far the sample statistic is from the null hypothesis, in standard units.

4. **Determine the p-value or critical value**

   * The **p-value** is the probability of observing a test statistic as extreme as, or more extreme than, the one observed if H‚ÇÄ is true.

5. **Make a Decision**

   * If **p-value ‚â§ Œ±**, reject the null hypothesis (evidence supports H‚ÇÅ).
   * If **p-value > Œ±**, fail to reject the null (insufficient evidence).

---

###  **Example**

Suppose you want to test whether a new drug is more effective than the standard treatment:

* H‚ÇÄ: The new drug has the same effectiveness as the old one.
* H‚ÇÅ: The new drug is more effective.

You collect data, run a t-test, and find a p-value of 0.03.
If Œ± = 0.05 ‚Üí Since 0.03 < 0.05, **reject H‚ÇÄ** ‚Üí Evidence suggests the new drug is more effective.

---

###  **Errors in Hypothesis Testing**

1. **Type I Error (False Positive)**

   * Rejecting H‚ÇÄ when it is actually true.
   * Controlled by the significance level Œ±.

2. **Type II Error (False Negative)**

   * Failing to reject H‚ÇÄ when it is actually false.
   * Probability is Œ≤ (power = 1 - Œ≤).

---




2.  What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The **null hypothesis (H‚ÇÄ)** and the **alternative hypothesis (H‚ÇÅ or Ha)** are the two foundational concepts in hypothesis testing. Here's a clear breakdown:

---

###  **Null Hypothesis (H‚ÇÄ)**

* **Definition**:
  A statement that there is **no effect, no difference, or no relationship** in the population. It represents the **status quo** or default position.

* **Purpose**:
  It is the hypothesis that we test against; we try to find evidence to **reject** it.

* **Example**:
  If a manufacturer claims that a battery lasts 10 hours, then:
  `H‚ÇÄ: Œº = 10` (the mean battery life is 10 hours)

---

###  **Alternative Hypothesis (H‚ÇÅ or Ha)**

* **Definition**:
  A statement that **contradicts** the null hypothesis. It suggests that **there is an effect, a difference, or a relationship**.

* **Purpose**:
  If the data provides strong enough evidence, we **accept this as the likely reality** by rejecting the null hypothesis.

* **Example**:
  If you're testing whether the battery lasts longer than 10 hours:
  `H‚ÇÅ: Œº > 10` (the mean battery life is more than 10 hours)

---

###  **Key Differences**

| Feature             | Null Hypothesis (H‚ÇÄ)                    | Alternative Hypothesis (H‚ÇÅ or Ha)     |
| ------------------- | --------------------------------------- | ------------------------------------- |
| Meaning             | No effect or no change                  | Some effect or change exists          |
| Assumption          | Accepted unless evidence says otherwise | Supported **if** H‚ÇÄ is rejected       |
| Goal in testing     | Try to **reject** it                    | Try to **support** it by rejecting H‚ÇÄ |
| Symbol              | H‚ÇÄ                                      | H‚ÇÅ or Ha                              |
| Example (mean test) | H‚ÇÄ: Œº = 50                              | H‚ÇÅ: Œº ‚â† 50 (or >, < 50)               |

---

###  Real-World Analogy

Think of **H‚ÇÄ as ‚Äúinnocent until proven guilty.‚Äù**
You assume innocence (H‚ÇÄ) unless there‚Äôs strong evidence (data) to convict (reject H‚ÇÄ and accept H‚ÇÅ).

---




3.  What is the significance level in hypothesis testing, and why is it important?
- ###  **Significance Level (Œ±) in Hypothesis Testing**

The **significance level**, denoted by **Œ± (alpha)**, is the **threshold** you set for deciding whether to reject the **null hypothesis (H‚ÇÄ)** in a statistical test.

---

###  **Definition**

* The **significance level (Œ±)** is the **maximum probability** of making a **Type I error** ‚Äî i.e., **rejecting the null hypothesis when it is actually true**.
* It represents how **strict or lenient** you are in detecting an effect.

---

###  **Common Significance Levels**

| Œ± Value | Meaning                               |
| ------- | ------------------------------------- |
| 0.05    | 5% risk of Type I error (most common) |
| 0.01    | 1% risk (more stringent)              |
| 0.10    | 10% risk (less stringent)             |

For example:
If Œ± = 0.05, you're willing to accept a **5% chance** of mistakenly rejecting H‚ÇÄ when it's actually true.

---

###  **Why is it Important?**

1. **Sets the Decision Boundary**

   * The significance level helps decide **how strong the evidence must be** to reject H‚ÇÄ.
   * If **p-value ‚â§ Œ±** ‚Üí reject H‚ÇÄ.
   * If **p-value > Œ±** ‚Üí fail to reject H‚ÇÄ.

2. **Controls False Alarms (Type I Errors)**

   * Lower Œ± = fewer false positives (more caution).
   * Higher Œ± = more sensitivity but higher risk of mistakes.

3. **Determines Critical Regions**

   * It defines the **critical value(s)** or rejection zone(s) in your test's probability distribution.

---

###  Example:

You're testing whether a coin is fair (H‚ÇÄ: p = 0.5). You set Œ± = 0.05.

* If your test yields a **p-value of 0.03**, then:

  * Since 0.03 < 0.05 ‚Üí **Reject H‚ÇÄ**
  * Evidence suggests the coin **may not be fair**.

---

###  Tip:

* Choose Œ± **before** conducting your test ‚Äî not after seeing the results.
* If the consequences of a false positive are serious (like in medical trials), choose a **smaller Œ±** (e.g., 0.01).

---




4.  What does a P-value represent in hypothesis testing?
- ###  **What is a P-value in Hypothesis Testing?**

The **p-value** (probability value) is a **measure of evidence** against the **null hypothesis (H‚ÇÄ)** in a statistical test.

---

###  **Definition**

> The **p-value** is the probability of obtaining a result **at least as extreme as the one observed**, **assuming the null hypothesis is true**.

---

###  **What It Tells You**

* **Low p-value (typically ‚â§ 0.05)**:
  ‚Üí Strong evidence **against** H‚ÇÄ ‚Üí **Reject** the null hypothesis.

* **High p-value (typically > 0.05)**:
  ‚Üí Weak evidence against H‚ÇÄ ‚Üí **Fail to reject** the null hypothesis.

---

###  **Example**

Suppose you test a new medicine:

* H‚ÇÄ: The drug has **no effect**.
* You get a p-value of **0.02**.

If your significance level Œ± = 0.05:

* Since 0.02 < 0.05 ‚Üí You **reject H‚ÇÄ**.
* There's only a 2% chance you'd see this result **if the drug really had no effect** ‚Üí Evidence **supports** the drug works.

---

###  **Key Points to Remember**

| Concept            | Explanation                                                               |
| ------------------ | ------------------------------------------------------------------------- |
| **P-value ‚â§ Œ±**    | Reject H‚ÇÄ (evidence supports H‚ÇÅ)                                          |
| **P-value > Œ±**    | Fail to reject H‚ÇÄ (insufficient evidence)                                 |
| **P-value is NOT** | The probability that H‚ÇÄ is true or false                                  |
| **P-value is NOT** | The probability that your result occurred by random chance **in general** |
| **P-value is**     | The probability of your result or more extreme **given that H‚ÇÄ is true**  |

---



5.  How do you interpret the P-value in hypothesis testing?
- ###  **How to Interpret the P-value in Hypothesis Testing**

The **p-value** helps you decide whether the sample data provides enough evidence to **reject the null hypothesis (H‚ÇÄ)**.

---

###  **General Interpretation**

| **P-value**             | **Interpretation**                                                |
| ----------------------- | ----------------------------------------------------------------- |
| **Small p-value (‚â§ Œ±)** | Strong evidence **against H‚ÇÄ** ‚Üí **Reject** the null hypothesis   |
| **Large p-value (> Œ±)** | Weak evidence against H‚ÇÄ ‚Üí **Fail to reject** the null hypothesis |

* **Œ±** is the **significance level** you choose (commonly 0.05 or 5%).

---

###  **Examples of Interpretation**

####  Scenario: Testing if a new teaching method improves student scores

**H‚ÇÄ**: The new method is **no better** than the traditional method.
**H‚ÇÅ**: The new method **improves** scores.

| P-value | Interpretation                                                                                     |
| ------- | -------------------------------------------------------------------------------------------------- |
| 0.03    | 3% chance of seeing this data **if H‚ÇÄ is true** ‚Üí **Reject H‚ÇÄ** (method likely improves scores)    |
| 0.08    | 8% chance of seeing this data under H‚ÇÄ ‚Üí **Fail to reject H‚ÇÄ** (no strong evidence of improvement) |

---

###  **Step-by-Step Use in Hypothesis Testing**

1. **State Hypotheses**

   * H‚ÇÄ: No effect/difference
   * H‚ÇÅ: There is an effect/difference

2. **Choose Significance Level (Œ±)**

   * Example: Œ± = 0.05

3. **Calculate Test Statistic and P-value**

4. **Compare P-value to Œ±**

   * If **p ‚â§ Œ±**: Reject H‚ÇÄ ‚Üí Evidence supports H‚ÇÅ
   * If **p > Œ±**: Fail to reject H‚ÇÄ ‚Üí Evidence is not strong enough

---

###  **Common Misunderstandings**

| Misconception                                | Reality                                                                       |
| -------------------------------------------- | ----------------------------------------------------------------------------- |
| ‚ÄúP-value = 0.01 means 99% chance H‚ÇÅ is true‚Äù |  No. It means there's a 1% chance of observing this result **if H‚ÇÄ is true** |
| ‚ÄúHigh p-value proves H‚ÇÄ is true‚Äù             |  No. It just means there's **not enough evidence to reject it**              |

---

###  Real-World Analogy

Think of **H‚ÇÄ as a ‚Äúdefault assumption,‚Äù** like a defendant being innocent.

* A **small p-value** is like strong evidence presented in court ‚Üí **Conviction** (reject H‚ÇÄ).
* A **large p-value** is like weak evidence ‚Üí **No conviction**, but doesn‚Äôt prove innocence (fail to reject H‚ÇÄ).

6.  What are Type 1 and Type 2 errors in hypothesis testing?
- ###  **Type I and Type II Errors in Hypothesis Testing**

When conducting hypothesis testing, there are two kinds of mistakes you can make:

---

###  **Type I Error (False Positive)**

* **Definition**: Rejecting the null hypothesis (**H‚ÇÄ**) when it is actually **true**.
* **Symbol**: Œ± (alpha) ‚Üí the **significance level**.
* **Example**:
  You conclude a new drug works better than the old one, but in reality, it doesn‚Äôt.

 **Think:** *"We saw an effect, but it wasn‚Äôt real."*

---

###  **Type II Error (False Negative)**

* **Definition**: Failing to reject the null hypothesis when it is actually **false**.
* **Symbol**: Œ≤ (beta)
* **Example**:
  You conclude the new drug doesn‚Äôt work better, but it actually **does**.

 **Think:** *"There was a real effect, but we missed it."*

---

###  **Visual Summary**

|                         | **Reality: H‚ÇÄ True** | **Reality: H‚ÇÄ False** |
| ----------------------- | -------------------- | --------------------- |
| **Decision: Reject H‚ÇÄ** |  Type I Error       |  Correct (Power)     |
| **Decision: Fail H‚ÇÄ**   |  Correct Decision   |  Type II Error       |

---

###  **Relation to Significance and Power**

* **Œ± (alpha)**: Probability of **Type I Error**
* **Œ≤ (beta)**: Probability of **Type II Error**
* **Power of the test** = 1 ‚àí Œ≤ ‚Üí The chance of correctly rejecting H‚ÇÄ when it is false

---

###  **Real-World Analogy**

Imagine a **courtroom**:

* **H‚ÇÄ**: The defendant is **innocent**
* **H‚ÇÅ**: The defendant is **guilty**

| Error Type    | Analogy                         |
| ------------- | ------------------------------- |
| Type I Error  | Convicting an innocent person   |
| Type II Error | Letting a guilty person go free |

---

###  **Minimizing Errors**

* Lowering **Œ±** reduces **Type I errors**, but may increase **Type II errors**.
* Increasing **sample size** helps reduce **both** errors and increases test **power**.

7.  What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- ###  **One-Tailed vs. Two-Tailed Tests in Hypothesis Testing**

The **difference** lies in **how** you define the **alternative hypothesis (H‚ÇÅ)** ‚Äî and **where** in the probability distribution you look for evidence against the **null hypothesis (H‚ÇÄ)**.

---

###  **1. One-Tailed Test**

####  **Definition**:

You test for an effect in **one direction only** (either greater than or less than a value).

####  Use When:

You have a **specific directional claim**.

####  Example:

You're testing if a drug **increases** recovery speed.

* **H‚ÇÄ**: Œº ‚â§ 10
* **H‚ÇÅ**: Œº > 10
  (Only testing **if it‚Äôs greater**, not less)

####  Focus:

* Only **one tail** of the distribution (either left or right)

####  Visual:

Imagine a bell curve ‚Äî you only shade **one side** (left or right).

---

###  **2. Two-Tailed Test**

####  **Definition**:

You test for an effect in **both directions** ‚Äî significantly higher **or** lower.

####  Use When:

You just want to test **for a difference**, not a direction.

####  Example:

You're testing if a machine produces rods that are **not exactly** 10 cm.

* **H‚ÇÄ**: Œº = 10
* **H‚ÇÅ**: Œº ‚â† 10
  (You're open to detecting if it‚Äôs **either too high or too low**)

####  Focus:

* **Both tails** of the distribution

####  Visual:

Imagine a bell curve ‚Äî you shade **both sides** (left and right extremes).

---

###  **Comparison Table**

| Feature                | One-Tailed Test                               | Two-Tailed Test                              |
| ---------------------- | --------------------------------------------- | -------------------------------------------- |
| Directionality         | One direction (>, <)                          | Both directions (‚â†)                          |
| Sensitivity            | More sensitive in one direction               | Less sensitive (divides Œ± across both tails) |
| Risk of Missing Effect | Higher if effect is in the opposite direction | Balanced for both sides                      |
| Example H‚ÇÅ             | Œº > Œº‚ÇÄ or Œº < Œº‚ÇÄ                              | Œº ‚â† Œº‚ÇÄ                                       |

---

###  Important Note:

If you're not **sure** about the direction of the effect beforehand, always use a **two-tailed test**.

8.  What is the Z-test, and when is it used in hypothesis testing?
- ###  **What is the Z-test in Hypothesis Testing?**

The **Z-test** is a statistical test used to determine **whether there is a significant difference between sample and population means**, or between two sample means, **when the population variance is known** and the data is approximately **normally distributed**.

---

###  **When to Use a Z-test**

Use the **Z-test** when **all of the following** conditions are met:

1. **The sample size is large** (typically **n ‚â• 30**)
2. **The population standard deviation (œÉ) is known**
3. **Data is approximately normally distributed** or the **Central Limit Theorem** applies
4. You're testing:

   * A population mean
   * A population proportion
   * The difference between two means or two proportions

---

###  **Common Types of Z-tests**

| Test Type                  | Use Case Example                                                          |
| -------------------------- | ------------------------------------------------------------------------- |
| **One-sample Z-test**      | Compare sample mean to a known population mean                            |
| **Two-sample Z-test**      | Compare means of two independent samples                                  |
| **Z-test for proportions** | Test if the proportion in a sample differs from the population proportion |

---

###  **Z-test Formula (One-Sample Mean)**

$$
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
$$

Where:

* $\bar{X}$ = sample mean
* $\mu$ = population mean (under H‚ÇÄ)
* $\sigma$ = population standard deviation
* $n$ = sample size

---

###  **Example**

Suppose a machine fills bottles with 500 ml of juice. You take a sample of 36 bottles and find:

* Sample mean = 495 ml
* Population œÉ = 12 ml
* Test whether the machine is underfilling at **Œ± = 0.05**

$$
Z = \frac{495 - 500}{12 / \sqrt{36}} = \frac{-5}{2} = -2.5
$$

Now look up the **Z critical value** or **p-value**:

* Since Z = -2.5 and Œ± = 0.05 (one-tailed), you **reject H‚ÇÄ** ‚Üí machine is likely underfilling.

---

###  **Z-test vs. T-test**

| Feature           | Z-test              | T-test                   |
| ----------------- | ------------------- | ------------------------ |
| Use when œÉ is...  | Known               | Unknown                  |
| Sample size       | Large (n ‚â• 30)      | Small (n < 30)           |
| Distribution used | Standard Normal (Z) | Student‚Äôs t-distribution |

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
- ###  **Z-score: Calculation and Interpretation in Hypothesis Testing**

---

###  **What is a Z-score?**

A **Z-score** (also called a **standard score**) tells you **how many standard deviations** a data point or sample statistic is from the **mean** of a distribution.

In **hypothesis testing**, it's used to measure **how far your sample result is from what the null hypothesis (H‚ÇÄ) expects**, assuming H‚ÇÄ is true.

---

###  **Z-score Formula**

#### üîπ For a **single value** from a population:

$$
Z = \frac{X - \mu}{\sigma}
$$

* $X$: observed value
* $\mu$: population mean
* $\sigma$: population standard deviation

---

#### üîπ For a **sample mean** (used in Z-test):

$$
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
$$

* $\bar{X}$: sample mean
* $\mu$: population mean under H‚ÇÄ
* $\sigma$: population standard deviation
* $n$: sample size

---

###  **What Does the Z-score Represent?**

| **Z-score Value** | **Interpretation**                                                 |              |                                                     |
| ----------------- | ------------------------------------------------------------------ | ------------ | --------------------------------------------------- |
| Z = 0             | Sample mean = population mean                                      |              |                                                     |
| Z = +1            | Sample mean is 1 standard deviation **above** the population mean  |              |                                                     |
| Z = -2            | Sample mean is 2 standard deviations **below** the population mean |              |                                                     |
|                   | Z                                                                  | > critical Z | Result is **statistically significant** ‚Üí reject H‚ÇÄ |

---

###  **Critical Z-values (for common Œ± levels)**

| Test Type      | Œ± = 0.05        | Œ± = 0.01      |
| -------------- | --------------- | ------------- |
| **Two-tailed** | ¬±1.96           | ¬±2.576        |
| **One-tailed** | 1.645 or -1.645 | 2.33 or -2.33 |

---

###  **Example**

You want to test whether students' average test score (sample) differs from a national average (Œº = 70).
Given:

* $\bar{X} = 74$,
* $\sigma = 8$,
* $n = 36$

$$
Z = \frac{74 - 70}{8 / \sqrt{36}} = \frac{4}{1.33} \approx 3.00
$$

* Z = 3.00 is **greater than 1.96** ‚Üí Significant at Œ± = 0.05 (two-tailed) ‚Üí **Reject H‚ÇÄ**.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
- ###  **What is the T-Distribution?**

The **T-distribution** (or **Student's t-distribution**) is a **probability distribution** used in **inferential statistics** when:

* The **sample size is small** (typically **n < 30**), **and**
* The **population standard deviation (œÉ)** is **unknown**

It‚Äôs **similar to the normal distribution**, but with **heavier tails** ‚Äî which means it accounts for **more variability** when sample sizes are small.

---

###  **Key Characteristics of the T-Distribution**

| Feature             | Description                                                             |
| ------------------- | ----------------------------------------------------------------------- |
| Shape               | Bell-shaped and symmetric (like the normal curve)                       |
| Tails               | **Thicker** than normal ‚Üí allows for more variability in small samples  |
| Mean                | Centered at 0                                                           |
| Spread              | Determined by **degrees of freedom (df = n ‚àí 1)**                       |
| Converges to Normal | As **sample size increases**, it **approaches the normal distribution** |

---

###  **T-distribution vs. Normal Distribution**

| Criteria            | **Use T-distribution**                                       | **Use Normal (Z) distribution** |
| ------------------- | ------------------------------------------------------------ | ------------------------------- |
| Population œÉ known? |  No                                                         |  Yes                           |
| Sample size         | Small (typically **n < 30**)                                 | Large (typically **n ‚â• 30**)    |
| Data distribution   | Approx. normal (especially important when n is small)        | Normal or large enough for CLT  |
| Use case            | Hypothesis testing or confidence intervals when œÉ is unknown | When œÉ is known or n is large   |

---

###  **Common Use Cases of the T-distribution**

* **One-sample t-test**: Compare sample mean to known value
* **Two-sample t-test**: Compare means from two independent groups
* **Paired t-test**: Compare means from the same group (before & after)
* **Confidence intervals**: For small samples with unknown œÉ

---

###  **Example**

You're testing whether a training program improves scores. You have:

* $\bar{X} = 82$,
* $\mu = 80$,
* $s = 5$,
* $n = 16$

Since **œÉ is unknown** and **n < 30**, use the **t-distribution**.

$$
t = \frac{\bar{X} - \mu}{s / \sqrt{n}} = \frac{82 - 80}{5 / \sqrt{16}} = \frac{2}{1.25} = 1.6
$$

Now compare with a **critical t-value** for **df = 15**, depending on your significance level (e.g., 2.131 for Œ± = 0.05 two-tailed).

* Since 1.6 < 2.131 ‚Üí **Fail to reject H‚ÇÄ**

11. What is the difference between a Z-test and a T-test?
- The **Z-test** and **T-test** are both statistical tests used to determine whether there is a significant difference between sample data and a population or between two sample groups. However, they are used in different situations based on the **sample size** and whether the **population standard deviation** is known.

---

###  Key Differences Between Z-test and T-test:

| Feature               | **Z-test**                                                                            | **T-test**                                                                                |
| --------------------- | ------------------------------------------------------------------------------------- | ----------------------------------------------------------------------------------------- |
| **Use Case**          | When population **standard deviation is known** and sample size is **large (n > 30)** | When population **standard deviation is unknown** and sample size is **small (n ‚â§ 30)**   |
| **Distribution Used** | Standard normal distribution (Z-distribution)                                         | Student‚Äôs t-distribution                                                                  |
| **Formula**           | $Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}$                                         | $t = \frac{\bar{X} - \mu}{s / \sqrt{n}}$                                                  |
| **Tail Thickness**    | Thinner tails (less variability in extreme values)                                    | Thicker tails (accounts for more variability with small samples)                          |
| **Assumptions**       | Population is normally distributed or n > 30; œÉ known                                 | Population is normally distributed; œÉ unknown                                             |
| **Examples of Use**   | Comparing a sample mean to a known population mean with known œÉ                       | Comparing a sample mean to a known population mean with **sample standard deviation (s)** |

---

###  When to Use Which?

* Use a **Z-test** if:

  * Sample size is large (n > 30)
  * Population standard deviation (œÉ) is known

* Use a **T-test** if:

  * Sample size is small (n ‚â§ 30)
  * Population standard deviation is unknown (you use the sample SD)

---

###  Types of T-tests:

* **One-sample t-test**: Compare sample mean to a population mean
* **Two-sample (independent) t-test**: Compare means of two independent samples
* **Paired t-test**: Compare means from the same group at different times or under different conditions

12.  What is the T-test, and how is it used in hypothesis testing?
-

A **T-test** is a **parametric statistical test** used to determine whether there is a significant difference between:

* A **sample mean** and a **population mean**, or
* The means of **two groups**.

It‚Äôs used when the **population standard deviation is unknown** and the **sample size is small (n ‚â§ 30)**.

---

###  Types of T-tests:

| T-test Type                         | Use Case                                                               |
| ----------------------------------- | ---------------------------------------------------------------------- |
| **One-sample T-test**               | Compare a single sample mean to a known population mean                |
| **Two-sample (independent) T-test** | Compare the means of two independent groups                            |
| **Paired T-test**                   | Compare means from the same group at two different times or conditions |

---

###  T-test in Hypothesis Testing

#### Step-by-step process:

1. **State Hypotheses**:

   * Null Hypothesis $H_0$: No difference (e.g., $\mu = \mu_0$ or $\mu_1 = \mu_2$)
   * Alternative Hypothesis $H_a$: There **is** a difference (e.g., $\mu \ne \mu_0$)

2. **Choose significance level** $\alpha$ (commonly 0.05)

3. **Calculate the test statistic**:

   $$
   t = \frac{\bar{X} - \mu}{s / \sqrt{n}}
   $$

   where:

   * $\bar{X}$ = sample mean
   * $\mu$ = population mean or comparison mean
   * $s$ = sample standard deviation
   * $n$ = sample size

4. **Find the degrees of freedom**: $df = n - 1$

5. **Compare t-value** with the **critical value** from the t-distribution table or use the **p-value** approach.

6. **Make a decision**:

   * If $|t| >$ critical t-value, or if p-value < Œ± ‚Üí **Reject $H_0$**
   * Otherwise ‚Üí **Fail to reject $H_0$**

---

###  Example:

Let‚Äôs say a company claims the average time to deliver a package is **2 days**. A sample of 10 deliveries shows a **mean of 2.3 days** with a **standard deviation of 0.4 days**. Is there evidence the mean delivery time is different?

**Steps**:

* $H_0: \mu = 2$, $H_a: \mu \ne 2$
* $\bar{X} = 2.3$, $s = 0.4$, $n = 10$
* $t = \frac{2.3 - 2}{0.4/\sqrt{10}} \approx 2.37$
* Degrees of freedom = 9
* Compare with critical t-value (‚âà2.262 at Œ± = 0.05 for df = 9)

Since **2.37 > 2.262**, **reject $H_0$**: there's a significant difference.

13.  What is the relationship between Z-test and T-test in hypothesis testing?
- ###  Relationship Between **Z-test** and **T-test** in Hypothesis Testing

The **Z-test** and **T-test** are closely related statistical tools used in **hypothesis testing** to determine whether a sample differs significantly from a population or another sample. Here's how they're related:

---

###  **1. Common Goal**

Both tests are used to:

* Test hypotheses about **population means**
* Determine if the observed difference is **statistically significant**

---

###  **2. Similar Structure**

Both use the general form of the test statistic:

$$
\text{Test Statistic} = \frac{\text{Sample Mean} - \text{Population Mean}}{\text{Standard Error}}
$$

* **Z-test formula**:

  $$
  Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
  $$

  (Uses **population standard deviation** $\sigma$)

* **T-test formula**:

  $$
  t = \frac{\bar{X} - \mu}{s / \sqrt{n}}
  $$

  (Uses **sample standard deviation** $s$)

---

###  **3. Dependence on Sample Size and œÉ (Standard Deviation)**

* Use **Z-test** when:

  * $n > 30$ (large sample size)
  * Population standard deviation $\sigma$ is known
* Use **T-test** when:

  * $n \leq 30$ (small sample size)
  * $\sigma$ is **unknown** (you estimate it with $s$)

---

###  **4. Distribution Used**

* **Z-test**: Standard Normal Distribution (mean = 0, SD = 1)
* **T-test**: Student‚Äôs T-distribution (heavier tails to account for variability in small samples)

As the **sample size increases**, the **T-distribution approaches the Z-distribution** ‚Äî this is a key link between the two:

>  **T-test becomes almost identical to the Z-test when n is large**.

---

###  Summary Table:

| Feature       | **Z-test**     | **T-test**                              |
| ------------- | -------------- | --------------------------------------- |
| Sample Size   | Large (n > 30) | Small (n ‚â§ 30)                          |
| œÉ Known?      | Yes            | No                                      |
| Distribution  | Normal (Z)     | T-distribution                          |
| Tail Behavior | Thinner        | Heavier (more conservative for small n) |
| Converges to  | ‚Äî              | Z-test as n ‚Üí ‚àû                         |

---

###  In Short:

**T-test is a generalized form of Z-test** that adjusts for the extra uncertainty when the population standard deviation is unknown. Both serve the same purpose but differ in assumptions and application.




14.  What is a confidence interval, and how is it used to interpret statistical results?
-

A **confidence interval** is a **range of values** that is likely to contain the **true population parameter** (like the mean or proportion) with a certain level of **confidence** (usually 95% or 99%).

---

###  In Simple Terms:

> A 95% confidence interval means:
> **"We are 95% confident that the true value lies within this interval."**

It **does not** mean there‚Äôs a 95% probability the true value is inside the interval ‚Äî instead, if you repeated the study 100 times, about **95 of those intervals** would contain the true population parameter.

---

###  Confidence Interval Formula for Mean (when œÉ is unknown):

$$
\bar{X} \pm t^* \left( \frac{s}{\sqrt{n}} \right)
$$

Where:

* $\bar{X}$ = sample mean
* $s$ = sample standard deviation
* $n$ = sample size
* $t^*$ = critical value from the t-distribution based on the confidence level and degrees of freedom

> If population standard deviation $\sigma$ is **known**, use a **Z-value** instead of $t^*$.

---

###  Example:

You measure the average weight of a sample of 25 people and get:

* Mean = 70 kg
* SD = 5 kg
* Confidence level = 95%

Then the **confidence interval** is:

$$
70 \pm t_{0.025, 24} \times \frac{5}{\sqrt{25}} = 70 \pm 2.064 \times 1 = (67.94, 72.06)
$$

 You‚Äôd interpret it as:

> ‚ÄúWe are 95% confident that the true average weight of the population is between 67.94 kg and 72.06 kg.‚Äù

---

###  How It Helps in Interpretation:

* If a **confidence interval does not contain the null hypothesis value** (e.g., population mean = 0), it suggests **statistical significance**.
* **Wider intervals** indicate more **uncertainty**, while **narrower intervals** indicate more **precision**.
* Often used alongside **p-values** in hypothesis testing for deeper understanding.

---

###  Relationship to Hypothesis Testing:

| Confidence Interval             | Hypothesis Testing Interpretation |
| ------------------------------- | --------------------------------- |
| Does **not** include null value | Reject null hypothesis            |
| **Includes** null value         | Fail to reject null hypothesis    |

---




15. What is the margin of error, and how does it affect the confidence interval?
-

The **margin of error** is the **maximum amount you're willing to be wrong** when estimating a population parameter (like the mean or proportion) based on a sample. It defines **how much the sample estimate could vary** from the true population value.

---

###  Formula:

For a **mean**, the margin of error is:

$$
\text{MoE} = z^* \cdot \left( \frac{\sigma}{\sqrt{n}} \right) \quad \text{(if } \sigma \text{ is known)}
$$

or

$$
\text{MoE} = t^* \cdot \left( \frac{s}{\sqrt{n}} \right) \quad \text{(if } \sigma \text{ is unknown)}
$$

Where:

* $z^*$ or $t^*$ = critical value based on confidence level (e.g., 1.96 for 95% confidence with Z)
* $\sigma$ or $s$ = standard deviation
* $n$ = sample size

---

###  Relationship to Confidence Interval:

The **confidence interval** is built around the sample statistic like this:

$$
\text{Confidence Interval} = \text{Point Estimate} \pm \text{Margin of Error}
$$

So, MoE directly determines the **width** of the interval:

* **Larger MoE** ‚Üí **Wider** interval ‚Üí more uncertainty
* **Smaller MoE** ‚Üí **Narrower** interval ‚Üí more precision

---

###  What Affects the Margin of Error?

| Factor                   | Effect on MoE                             |
| ------------------------ | ----------------------------------------- |
| **Confidence level ‚Üë**   | MoE ‚Üë (wider interval, more confidence)   |
| **Sample size ‚Üë**        | MoE ‚Üì (narrower interval, more precision) |
| **Standard deviation ‚Üë** | MoE ‚Üë (more variability in the data)      |

---

###  Example:

Suppose:

* Sample mean = 50
* Standard deviation = 10
* Sample size = 100
* Confidence level = 95% (z\* = 1.96)

Then:

$$
\text{MoE} = 1.96 \cdot \left( \frac{10}{\sqrt{100}} \right) = 1.96 \cdot 1 = 1.96
$$

$$
\text{CI} = 50 \pm 1.96 = (48.04, 51.96)
$$

So the **true mean** is likely between 48.04 and 51.96 with **95% confidence**.

---

###  Summary:

* The **margin of error** is the "wiggle room" around your sample result.
* It helps create the **confidence interval**, showing how close your estimate might be to the truth.
* A **smaller margin of error** = more precise results.




16.  How is Bayes' Theorem used in statistics, and what is its significance?
-

**Bayes' Theorem** is a fundamental concept in probability and statistics that allows us to **update the probability of a hypothesis** based on new evidence.

It answers this question:

> ‚ÄúGiven new data, how should I update what I believe about a situation?‚Äù

---

###  Bayes' Theorem Formula:

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

Where:

* $P(A|B)$ = Posterior probability (Probability of **A given B** ‚Äî **updated belief**)
* $P(B|A)$ = Likelihood (Probability of **B if A is true**)
* $P(A)$ = Prior probability (Belief about A **before seeing B**)
* $P(B)$ = Marginal probability (Total probability of B)

---

###  In Simple Terms:

> **Posterior = (Likelihood √ó Prior) / Evidence**

* **Prior**: What you believed before seeing the data
* **Likelihood**: How likely is the data assuming your belief is true
* **Posterior**: What you believe after seeing the data

---

###  Example: Medical Testing

Suppose:

* A rare disease affects 1 in 1,000 people ‚Üí $P(Disease) = 0.001$
* The test is 99% accurate:

  * True Positive Rate (Sensitivity): $P(Pos|Disease) = 0.99$
  * False Positive Rate: $P(Pos|No Disease) = 0.01$

You take the test and it‚Äôs **positive**. What‚Äôs the chance you actually have the disease?

---

#### Using Bayes‚Äô Theorem:

Let:

* A = Having the disease
* B = Testing positive

$$
P(Disease|Positive) = \frac{P(Positive|Disease) \cdot P(Disease)}{P(Positive)}
$$

Where:

$$
P(Positive) = (P(Positive|Disease) \cdot P(Disease)) + (P(Positive|No Disease) \cdot P(No Disease))
$$

$$
= (0.99 \cdot 0.001) + (0.01 \cdot 0.999) = 0.00099 + 0.00999 = 0.01098
$$

$$
P(Disease|Positive) = \frac{0.99 \cdot 0.001}{0.01098} \approx 0.09 \text{ or } 9\%
$$

> Even after a **positive test**, the chance of having the disease is only **9%** ‚Äî because the disease is rare and false positives matter.

---

###  Significance of Bayes' Theorem in Statistics:

| Feature                                   | Importance                                                                        |
| ----------------------------------------- | --------------------------------------------------------------------------------- |
| **Updates beliefs**                       | Helps revise probabilities as new data becomes available                          |
| **Forms the basis of Bayesian inference** | Key to Bayesian statistics where parameters are treated as random variables       |
| **Used in decision-making**               | Especially when outcomes are uncertain and new evidence keeps coming              |
| **Real-world applications**               | Spam filtering, medical diagnosis, fraud detection, machine learning, A/B testing |

---

###  Frequentist vs. Bayesian:

| Approach        | Treats Probability As...     | Key Difference                         |
| --------------- | ---------------------------- | -------------------------------------- |
| **Frequentist** | Long-run frequency of events | Focuses on **data**                    |
| **Bayesian**    | Degree of belief             | Combines **prior belief** and **data** |

---

###  Summary:

Bayes‚Äô Theorem is crucial for **reasoning under uncertainty**, allowing us to blend **prior knowledge** with **new evidence** to make better statistical decisions.



17. What is the Chi-square distribution, and when is it used?
-

The **Chi-square distribution** is a **right-skewed**, continuous probability distribution that is **used in inferential statistics**, especially when dealing with **categorical data**.

It is **formed by summing the squares of independent standard normal variables**:

$$
\chi^2 = Z_1^2 + Z_2^2 + \cdots + Z_k^2
$$

Where $Z_i \sim N(0, 1)$, and $k$ is the **degrees of freedom (df)**.

---

###  Key Properties:

| Feature  | Description                                              |
| -------- | -------------------------------------------------------- |
| Shape    | Right-skewed (becomes more normal as df increases)       |
| Domain   | Only takes **non-negative values** (œá¬≤ ‚â• 0)              |
| Mean     | Equal to **degrees of freedom** (df)                     |
| Variance | Equal to **2 √ó df**                                      |
| Skewness | Higher for smaller df; approaches normal as df increases |

---

###  **When Is the Chi-square Distribution Used?**

####  1. **Chi-square Test of Independence**

Used to test **whether two categorical variables are related**.

* Example: Is **gender** independent of **voting preference**?

####  2. **Chi-square Goodness-of-Fit Test**

Used to test **if a sample matches an expected distribution**.

* Example: Do dice rolls follow a **uniform distribution**?

####  3. **Chi-square Test for Variance**

Used to test hypotheses about the **variance or standard deviation** of a normally distributed population.

####  4. **In Confidence Intervals for Variance**

For example, confidence intervals for the **variance of a population** use the chi-square distribution.

---

###  Chi-square Test Statistic Formula:

For tests with **observed vs. expected frequencies**:

$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
$$

Where:

* $O_i$ = Observed frequency
* $E_i$ = Expected frequency

---

###  Example: Chi-square Test of Independence

You survey 100 people on whether they prefer coffee or tea and note gender. You want to know if **drink preference depends on gender**.

* Use a **contingency table**
* Calculate expected frequencies
* Use the formula above to compute $\chi^2$
* Compare with critical value or p-value at desired significance level (e.g., 0.05)

---

###  Visual Insight:

* With **low degrees of freedom**, œá¬≤ is very skewed.
* As **df increases**, the curve becomes more symmetric, resembling a **normal distribution**.

---

###  Summary:

| Aspect   | Chi-square Distribution                                  |
| -------- | -------------------------------------------------------- |
| Type     | Continuous, non-negative                                 |
| Used For | Categorical data tests (independence, goodness-of-fit)   |
| Shape    | Right-skewed, more normal with higher df                 |
| Key in   | Hypothesis testing and confidence intervals for variance |

18. What is the Chi-square goodness of fit test, and how is it applied?
-

The **Chi-square goodness-of-fit test** is a **statistical test** used to determine **whether a set of observed categorical data matches an expected distribution**.

> It helps you answer:
> **‚ÄúDo my observed frequencies differ significantly from what I expected?‚Äù**

---

###  When to Use:

* You have **1 categorical variable**.
* You want to check if the **observed frequencies** match **theoretical expectations**.

---

###  Test Statistic Formula:

$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
$$

Where:

* $O_i$ = observed frequency for category $i$
* $E_i$ = expected frequency for category $i$

---

###  Steps to Apply the Chi-square Goodness-of-Fit Test:

#### üîπ Step 1: **State the hypotheses**

* **Null (H‚ÇÄ):** Observed distribution fits the expected distribution
* **Alternative (H‚ÇÅ):** Observed distribution does **not** fit the expected distribution

---

#### üîπ Step 2: **Calculate Expected Frequencies**

* If the distribution is uniform, $E_i = \frac{\text{Total Observations}}{\text{Number of Categories}}$
* If probabilities differ, use $E_i = P_i \times \text{Total Observations}$

---

#### üîπ Step 3: **Compute the Chi-square Statistic**

$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
$$

---

#### üîπ Step 4: **Find Degrees of Freedom**

$$
df = \text{Number of categories} - 1
$$

---

#### üîπ Step 5: **Compare with Critical Value or Compute p-value**

* Use the **Chi-square distribution table** or software
* If $\chi^2_{\text{calc}} > \chi^2_{\text{critical}}$ or p-value < Œ± ‚Üí **Reject H‚ÇÄ**

---

###  Example:

You roll a 6-sided die 60 times and observe the outcomes:

| Face | Observed (O) |
| ---- | ------------ |
| 1    | 8            |
| 2    | 10           |
| 3    | 9            |
| 4    | 12           |
| 5    | 11           |
| 6    | 10           |

Assuming a **fair die**, expected frequency per face:

$$
E = \frac{60}{6} = 10
$$

$$
\chi^2 = \sum \frac{(O - E)^2}{E} = \frac{(8-10)^2}{10} + \cdots + \frac{(10-10)^2}{10} = 0.4 + 0 + 0.1 + 0.4 + 0.1 + 0 = 1.0
$$

* $df = 6 - 1 = 5$
* Critical value at Œ± = 0.05 and df = 5 is **11.07**

Since $\chi^2 = 1.0 < 11.07$, **do not reject** H‚ÇÄ.

 **Conclusion**: The die appears fair.

---

###  Assumptions:

* Categories are **mutually exclusive**
* Data are **frequencies**, not percentages or continuous values
* Expected frequency in each category should be **‚â• 5** (preferably)

---

###  Summary Table:

| Feature        | Chi-square Goodness-of-Fit Test           |
| -------------- | ----------------------------------------- |
| Purpose        | Compare observed vs. expected frequencies |
| Data type      | Categorical                               |
| Hypotheses     | H‚ÇÄ: Observed = Expected                   |
| Test statistic | $\chi^2 = \sum \frac{(O - E)^2}{E}$       |
| Decision rule  | Compare with critical value or p-value    |
| Assumptions    | Frequencies, independence, expected ‚â• 5   |

19. What is the F-distribution, and when is it used in hypothesis testing?
-

The **F-distribution** is a **right-skewed**, continuous probability distribution that arises when comparing **two variances**. It‚Äôs most commonly used in **hypothesis testing** involving **ratios of variances**.

It is defined as:

$$
F = \frac{S_1^2 / \text{df}_1}{S_2^2 / \text{df}_2}
$$

Where:

* $S_1^2$, $S_2^2$ are sample variances
* $\text{df}_1$, $\text{df}_2$ are degrees of freedom for the respective variances

---

###  Key Characteristics:

| Property      | Description                                                                    |
| ------------- | ------------------------------------------------------------------------------ |
| Shape         | Right-skewed, becomes more symmetric with higher degrees of freedom            |
| Range         | Only positive values (F ‚â• 0)                                                   |
| Parameters    | Two degrees of freedom: $\text{df}_1$ (numerator), $\text{df}_2$ (denominator) |
| Not symmetric | Used only for one-tailed tests (typically)                                     |

---

###  When is the F-distribution used?

####  1. **ANOVA (Analysis of Variance)**

> **Most common application**

Used to compare **three or more means** by analyzing **between-group vs. within-group variance**.

* If the **between-group variance** is significantly greater than the **within-group variance**, at least one group mean is different.

####  2. **Test of Equality of Variances**

Used to check if **two populations have equal variances**, before performing a t-test.

####  3. **Regression Analysis**

* To test whether the **overall regression model is significant**
* Based on the ratio of **regression variance to residual variance**

---

###  Example (ANOVA):

You test the effectiveness of 3 teaching methods on student scores.

| Group | Sample Mean | Sample Variance |
| ----- | ----------- | --------------- |
| A     | 70          | 9               |
| B     | 75          | 12              |
| C     | 80          | 15              |

* You use ANOVA to compare whether the **means are significantly different**.
* Compute **F = (variance between groups) / (variance within groups)**
* Compare the **F-value to the F-critical value** at a given significance level (e.g., Œ± = 0.05)

---

###  Summary Table:

| Feature   | Description                                             |
| --------- | ------------------------------------------------------- |
| Purpose   | Compare variances or test multiple means (ANOVA)        |
| Statistic | Ratio of two variances                                  |
| Values    | Positive only                                           |
| Tests     | ANOVA, variance equality, regression model significance |
| Shape     | Right-skewed, depends on df‚ÇÅ and df‚ÇÇ                    |

---

###  Visual:

* At low degrees of freedom: highly skewed
* As df increases: curve becomes smoother and approaches normal shape

20.  What is an ANOVA test, and what are its assumptions?
-

**ANOVA** stands for **Analysis of Variance**. It's a statistical method used to determine whether **three or more group means** are **significantly different** from each other.

> It tells you if **at least one group mean** is different, but **not which one** ‚Äî for that, post-hoc tests like Tukey's HSD are used.

---

###  Why Use ANOVA?

When comparing means:

* 2 groups ‚Üí Use a **t-test**
* 3 or more groups ‚Üí Use **ANOVA**

Using multiple t-tests increases the **Type I error rate** (false positives), while ANOVA keeps the error rate controlled.

---

###  ANOVA Hypotheses:

* **Null Hypothesis (H‚ÇÄ):** All group means are equal

  $$
  \mu_1 = \mu_2 = \mu_3 = \cdots = \mu_k
  $$
* **Alternative Hypothesis (H‚ÇÅ):** At least one group mean is different

---

###  ANOVA Test Statistic:

$$
F = \frac{\text{Between-group variance}}{\text{Within-group variance}} = \frac{MS_{\text{between}}}{MS_{\text{within}}}
$$

Where:

* $MS_{\text{between}}$ = Mean square between groups
* $MS_{\text{within}}$ = Mean square within groups
* Higher F = more likely at least one mean is different

---

###  Types of ANOVA:

| Type                        | Use Case                                                              |
| --------------------------- | --------------------------------------------------------------------- |
| **One-Way ANOVA**           | Compare means across **one factor** (e.g., teaching method)           |
| **Two-Way ANOVA**           | Compare means across **two factors** (e.g., teaching method & gender) |
| **Repeated Measures ANOVA** | Used when **same subjects** are tested multiple times                 |

---

###  Assumptions of ANOVA:

| Assumption                  | Description                                                                                |
| --------------------------- | ------------------------------------------------------------------------------------------ |
| **Independence**            | Observations in each group must be **independent**                                         |
| **Normality**               | The dependent variable should be **approximately normally distributed** in each group      |
| **Homogeneity of variance** | All groups should have **equal variances** (tested using Levene‚Äôs Test or Bartlett‚Äôs Test) |

---

###  Example:

You want to compare test scores of students taught with 3 different teaching methods:

| Group | Mean Score |
| ----- | ---------- |
| A     | 70         |
| B     | 75         |
| C     | 85         |

Use **One-Way ANOVA** to test whether the **difference in means** is statistically significant.

---

###  Summary:

| Feature         | Description                                        |
| --------------- | -------------------------------------------------- |
| What it does    | Tests for **differences in means** among 3+ groups |
| Test statistic  | F-ratio (variance between / variance within)       |
| Null hypothesis | All group means are equal                          |
| Output          | F-value and p-value                                |
| If p < Œ±        | At least one group is different (reject H‚ÇÄ)        |

21. What are the different types of ANOVA tests?
-
### 1. **One-Way ANOVA**

* **Purpose:** Compare means of **3 or more groups** based on **one independent factor**.

* **Assumptions:** One categorical independent variable, continuous dependent variable.

---

* **Output:** Tests main effects of each factor plus their interaction effect.

* **Purpose:** Combines **between-subjects factors** and **within-subjects factors**.
* **Example:** Testing different diets (between-subjects) and measuring weight loss at multiple times (within-subjects).
* Allows testing interaction between the two types of factors.

---

### 5. **MANOVA (Multivariate ANOVA)**

* **Purpose:** Extension of ANOVA to test differences in **multiple dependent variables** simultaneously.
* **Example:** Comparing groups on multiple outcomes like blood pressure and cholesterol levels together.
* Accounts for correlations between dependent variables.

---

22.  What is the F-test, and how does it relate to hypothesis testing?
-

The **F-test** is a statistical test that uses the **F-distribution** to compare **two variances** or to test the overall significance of a model (like in ANOVA or regression).

---

###  What Does the F-test Do?

* It measures the **ratio of variances**:

  $$
  F = \frac{\text{Variance estimate 1}}{\text{Variance estimate 2}}
  $$

* Larger values of $F$ indicate a greater difference between variances or groups.

---

###  Common Uses of the F-test in Hypothesis Testing:

#### 1. **Comparing Two Variances**

* Test if two population variances are equal.
* Hypotheses:

  * $H_0: \sigma_1^2 = \sigma_2^2$
  * $H_1: \sigma_1^2 \neq \sigma_2^2$
* Use ratio of sample variances, compare to F critical value.

#### 2. **ANOVA (Analysis of Variance)**

* Test if **means across multiple groups** are equal.
* Hypotheses:

  * $H_0: \mu_1 = \mu_2 = \cdots = \mu_k$
  * $H_1:$ At least one group mean differs.
* Uses ratio of **between-group variance** to **within-group variance**.
* $F = \frac{MS_{\text{between}}}{MS_{\text{within}}}$

#### 3. **Regression Analysis**

* Test whether the overall regression model is statistically significant.
* Hypotheses:

  * $H_0:$ No relationship (all coefficients zero except intercept).
  * $H_1:$ At least one predictor is significant.
* Compares model variance explained to unexplained variance.

---

###  How Does It Relate to Hypothesis Testing?

* The **F-statistic** measures the relative amount of explained vs. unexplained variance.
* You calculate the F-value from your data and compare it to the critical F-value from the F-distribution table with corresponding degrees of freedom.
* If $F_{\text{calculated}} > F_{\text{critical}}$ (or p-value < Œ±), you **reject the null hypothesis**.
* This means there's statistically significant evidence of differences or relationships.

---

### Summary Table:

| Aspect              | Description                                            |
| ------------------- | ------------------------------------------------------ |
| Purpose             | Compare variances or group means                       |
| Test statistic      | Ratio of variances $F = \frac{variance_1}{variance_2}$ |
| Distribution        | F-distribution (depends on two dfs)                    |
| Null hypothesis     | Variances equal or means equal                         |
| Decision rule       | Reject $H_0$ if $F_{calc} > F_{crit}$ or p < Œ±         |
| Common applications | Variance comparison, ANOVA, regression                 |

````python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
#uniform distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

def plot_discrete_uniform_cdf(low, high):
    """
    Calculate and plot the CDF of a discrete uniform distribution
    with integer support from `low` (inclusive) to `high` (exclusive).
    """
    # Support values
    x = np.arange(low, high + 1)  # include high for plotting a full step at the end

    # Theoretical CDF values using scipy.stats.randint
    cdf_values = randint.cdf(x, low, high)

    # Plot using a step function
    plt.step(x, cdf_values, where='post')
    plt.ylim(-0.05, 1.05)
    plt.xlim(low - 1, high + 1)
    plt.title(f'CDF of Discrete Uniform Distribution [{low}, {high-1}]')
    plt.xlabel('Outcome')
    plt.ylabel('Cumulative Probability')
    plt.xticks(np.arange(low, high))
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

if __name__ == "__main__":
    # Example: uniform over {1, 2, 3, 4, 5, 6}
    plot_discrete_uniform_cdf(low=1, high=7)
````

````python
#7. Generate a continuous uniform distribution using NumPy and visualize it
import numpy as np
import matplotlib.pyplot as plt

def plot_continuous_uniform(a, b, num_samples=10000, num_bins=50):
    """
    Generate samples from a continuous uniform distribution on [a, b]
    and plot the histogram alongside the theoretical PDF.
    """
    # 1. Generate samples
    samples = np.random.uniform(low=a, high=b, size=num_samples)

    # 2. Empirical histogram (normalized to form a density)
    counts, bins, patches = plt.hist(
        samples,
        bins=num_bins,
        range=(a, b),
        density=True,
        edgecolor='black',
        alpha=0.7,
        label='Empirical histogram'
    )

    # 3. Theoretical PDF: f(x) = 1/(b - a) for x in [a, b]
    x = np.linspace(a, b, 200)
    pdf = np.ones_like(x) * (1.0 / (b - a))
    plt.plot(x, pdf, linewidth=2, label=f'Theoretical PDF (1/({b}-{a}))')

    # 4. Decorations
    plt.title(f'Continuous Uniform Distribution on [{a}, {b}]')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    # Example: uniform on [0, 1]
    plot_continuous_uniform(a=0.0, b=1.0)
````

````python
#9.  Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a 1D numerical dataset.

    Parameters:
        data (array-like): List or array of numbers.

    Returns:
        np.ndarray: Z-scores for each element in `data`.
    """
    data = np.asarray(data, dtype=float)
    mean = np.mean(data)
    std = np.std(data)
    if std == 0:
        raise ValueError("Standard deviation is zero; Z-scores undefined.")
    return (data - mean) / std

def plot_z_scores(data):
    """
    Compute and plot the Z-scores of a dataset.

    Parameters:
        data (array-like): List or array of numbers.

    Returns:
        np.ndarray: The computed Z-scores.
    """
    z = calculate_z_scores(data)
    idx = np.arange(len(z))

    # stem plot without deprecated arguments
    plt.stem(idx, z, basefmt=" ")
    plt.axhline(0, linewidth=1)
    plt.title("Z-Scores of the Dataset")
    plt.xlabel("Index")
    plt.ylabel("Z-score")
    # only put ticks on a subset if too many points
    if len(idx) <= 20:
        plt.xticks(idx)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    return z

if __name__ == "__main__":
    # Example usage:
    sample_data = [5, 7, 8, 2, 10, 3, 6, 9]
    z_scores = plot_z_scores(sample_data)
    print("Z-scores:", z_scores)

````

````python
#10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def clt_demo_exponential(n=30, num_experiments=10000, scale=1.0):
    """
    Demonstrate the Central Limit Theorem by sampling from an exponential distribution
    and plotting the distribution of sample means alongside its normal approximation.

    Parameters:
        n (int):               Sample size for each experiment.
        num_experiments (int): Number of repeated experiments.
        scale (float):         Scale (1/Œª) parameter of the exponential distribution.
    """
    # 1. Draw num_experiments samples of size n from Exp(scale)
    data = np.random.exponential(scale=scale, size=(num_experiments, n))

    # 2. Compute sample means
    sample_means = data.mean(axis=1)

    # 3. Theoretical mean and std-dev of the sampling distribution
    Œº = scale
    œÉ = scale / np.sqrt(n)

    # 4. Plot histogram of the sample means
    plt.hist(sample_means,
             bins=30,
             density=True,
             edgecolor='black',
             alpha=0.7,
             label='Sample means')

    # 5. Overlay the corresponding normal PDF
    x = np.linspace(sample_means.min(), sample_means.max(), 300)
    plt.plot(x,
             norm.pdf(x, loc=Œº, scale=œÉ),
             linewidth=2,
             label='Normal approx.')

    # 6. Decorations
    plt.title(f'CLT Demo: Exponential (n={n}, experiments={num_experiments})')
    plt.xlabel('Sample mean')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    clt_demo_exponential(n=30, num_experiments=10000, scale=1.0)
````

````python
#11. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def clt_demo_normal(mu=0, sigma=1, n=30, num_experiments=10000):
    """
    Demonstrate the Central Limit Theorem by sampling from a normal distribution
    and plotting the distribution of sample means alongside its theoretical normal PDF.

    Parameters:
        mu (float):             Mean of the original normal distribution.
        sigma (float):          Standard deviation of the original normal distribution.
        n (int):                Sample size for each experiment.
        num_experiments (int):  Number of repeated experiments.
    """
    # 1. Draw num_experiments samples of size n from N(mu, sigma^2)
    data = np.random.normal(loc=mu, scale=sigma, size=(num_experiments, n))

    # 2. Compute sample means
    sample_means = data.mean(axis=1)

    # 3. Theoretical mean and std-dev of the sampling distribution
    theoretical_mean = mu
    theoretical_std = sigma / np.sqrt(n)

    # 4. Plot histogram of the sample means
    plt.hist(sample_means, bins=30, density=True, edgecolor='black', alpha=0.7, label='Sample means')

    # 5. Overlay the theoretical normal PDF
    x = np.linspace(sample_means.min(), sample_means.max(), 300)
    plt.plot(x, norm.pdf(x, loc=theoretical_mean, scale=theoretical_std), linewidth=2, label='Normal approx.')

    # 6. Decorations
    plt.title(f'CLT Demo (Normal): Œº={mu}, œÉ={sigma}, n={n}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    clt_demo_normal(mu=0, sigma=1, n=30, num_experiments=10000)
````

````python
#13.  Generate random variables and calculate their corresponding probabilities using the binomial distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def simulate_binomial_and_compute_probs(n, p, num_samples=10000):
    """
    1. Generate `num_samples` random variables from a Binomial(n, p) distribution.
    2. Compute the empirical probability of each outcome.
    3. Compute the theoretical probability of each outcome via the Binomial PMF.
    4. Plot the empirical histogram and overlay the theoretical PMF.

    Parameters:
        n (int):             Number of trials in each binomial draw.
        p (float):           Probability of success on each trial.
        num_samples (int):   How many binomial draws to simulate.

    Returns:
        values (ndarray):            Sorted array of outcomes observed (0 through n).
        empirical_probs (ndarray):   Empirical probabilities from the simulation.
        theoretical_probs (ndarray): Theoretical PMF values for each outcome.
    """
    # 1. Simulate
    samples = np.random.binomial(n, p, size=num_samples)

    # 2. Empirical probabilities
    values, counts = np.unique(samples, return_counts=True)
    empirical_probs = counts / num_samples

    # 3. Theoretical probabilities
    theoretical_probs = binom.pmf(values, n, p)

    # 4. Print a comparison table
    print(f"{'k':>3} | {'Empirical PÃÇ(X=k)':>18} | {'Theoretical P(X=k)':>20}")
    print("-" * 50)
    for k, emp, theo in zip(values, empirical_probs, theoretical_probs):
        print(f"{k:>3} | {emp:>18.4f} | {theo:>20.4f}")

    # 5. Plot
    plt.bar(values - 0.2, empirical_probs,
            width=0.4, label="Empirical",
            edgecolor='black', alpha=0.7)
    plt.bar(values + 0.2, theoretical_probs,
            width=0.4, label="Theoretical PMF",
            edgecolor='black', alpha=0.7)
    plt.title(f"Binomial Distribution (n={n}, p={p})")
    plt.xlabel("Number of Successes k")
    plt.ylabel("Probability")
    plt.xticks(np.arange(0, n+1))
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    return values, empirical_probs, theoretical_probs

if __name__ == "__main__":
    # Example usage:
    n = 10
    p = 0.3
    simulate_binomial_and_compute_probs(n, p, num_samples=10000)
````

````python
#14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
#distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, data):
    """
    Calculate the Z-score of a single data point x given a dataset.

    Parameters:
        x (float):      The data point.
        data (array-like): The dataset (list or array of numbers).

    Returns:
        float: The Z-score of x.
    """
    data = np.asarray(data, dtype=float)
    Œº = data.mean()
    œÉ = data.std(ddof=0)  # population std
    if œÉ == 0:
        raise ValueError("Standard deviation of the dataset is zero.")
    return (x - Œº) / œÉ

def compare_to_standard_normal(z):
    """
    Given a Z-score, compute its cumulative probability P(Z <= z) and
    plot the standard normal PDF marking the point z.

    Parameters:
        z (float): The Z-score.

    Returns:
        float: cumulative probability P(Z <= z).
    """
    # CDF at z
    p = norm.cdf(z)

    # Plot standard normal PDF
    x = np.linspace(-4, 4, 400)
    pdf = norm.pdf(x)

    plt.plot(x, pdf, linewidth=2, label='Standard Normal PDF')
    plt.fill_between(x, pdf, where=(x <= z), alpha=0.3,
                     label=f'P(Z ‚â§ {z:.2f}) = {p:.4f}')

    plt.axvline(z, color='red', linestyle='--', label=f'z = {z:.2f}')
    plt.title('Standard Normal Distribution')
    plt.xlabel('Z')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    return p

if __name__ == "__main__":
    # Example dataset
    data = [12, 15, 14, 10, 18, 20, 11, 13, 16, 14]

    # Data point to evaluate
    x = 17

    # 1. Calculate Z-score
    z = calculate_z_score(x, data)
    print(f"Z-score of x = {x} is {z:.4f}")

    # 2. Compare to standard normal
    p = compare_to_standard_normal(z)
    print(f"Cumulative probability P(Z ‚â§ {z:.4f}) = {p:.4f}")
````

````python
#15. Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

def z_test(sample, popmean, popstd, alternative='two-sided'):
    """
    Perform a one-sample Z-test.

    Parameters:
        sample (array-like):       Observed data.
        popmean (float):           Hypothesized population mean (H0: Œº = popmean).
        popstd (float):            Known population standard deviation œÉ.
        alternative (str):         'two-sided', 'larger', or 'smaller'.
                                   - 'two-sided': H1: Œº ‚â† popmean
                                   - 'larger':    H1: Œº > popmean
                                   - 'smaller':   H1: Œº < popmean

    Returns:
        z_stat (float):            Z-test statistic.
        p_value (float):           Corresponding p-value.
    """
    sample = np.asarray(sample, dtype=float)
    n = sample.size
    xbar = sample.mean()

    # Compute Z statistic
    z_stat = (xbar - popmean) / (popstd / np.sqrt(n))

    # Compute p-value
    if alternative == 'two-sided':
        p_value = 2 * norm.sf(abs(z_stat))
    elif alternative == 'larger':
        # P(Z >= z_stat)
        p_value = norm.sf(z_stat)
    elif alternative == 'smaller':
        # P(Z <= z_stat)
        p_value = norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-sided', 'larger', or 'smaller'")

    return z_stat, p_value

if __name__ == "__main__":
    # Example dataset
    sample_data = [5.1, 4.9, 5.0, 5.2, 4.8, 5.3, 5.1, 4.7, 5.0, 5.2]

    # Known population parameters under H0
    hypothesized_mean = 5.0
    known_std = 0.3   # assume œÉ known

    # Perform two-sided Z-test
    z_stat, p_val = z_test(sample_data, hypothesized_mean, known_std, alternative='two-sided')

    print(f"Sample mean = {np.mean(sample_data):.3f}")
    print(f"Z statistic = {z_stat:.3f}")
    print(f"P-value (two-sided) = {p_val:.4f}")

    # Interpret
    alpha = 0.05
    if p_val < alpha:
        print("Reject H0 at Œ± =", alpha)
    else:
        print("Fail to reject H0 at Œ± =", alpha)
````

````python
#16. Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Compute a two-sided confidence interval for the mean of `data`
    using the Student‚Äôs t-distribution.

    Parameters:
        data (array-like):     Your sample observations.
        confidence (float):     Desired confidence level (0 < confidence < 1).

    Returns:
        (mean, lower_bound, upper_bound)
    """
    data = np.asarray(data, dtype=float)
    n = len(data)
    if n < 2:
        raise ValueError("Need at least two data points to compute a confidence interval.")

    mean = np.mean(data)
    sem = stats.sem(data)  # standard error of the mean:  s / sqrt(n)

    # t critical value for two-sided interval
    alpha = 1.0 - confidence
    df = n - 1
    t_crit = stats.t.ppf(1.0 - alpha/2, df)

    margin = t_crit * sem
    return mean, mean - margin, mean + margin

if __name__ == "__main__":
    # Example dataset
    sample_data = [12.1, 13.3, 11.8, 12.7, 13.0, 12.4, 12.9, 13.2, 11.9, 12.5]

    # Compute 95% confidence interval
    mean, lower, upper = confidence_interval(sample_data, confidence=0.95)

    print(f"Sample mean            = {mean:.3f}")
    print(f"95% confidence interval = [{lower:.3f}, {upper:.3f}]")

    # Interpretation
    print("\nInterpretation:")
    print(f"We are 95% confident that the true population mean lies between {lower:.3f} and {upper:.3f}.")
````

````python
#17.  Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def simulate_and_ci(mu_true=5.0, sigma_true=2.0, n=50, confidence=0.95):
    """
    1. Simulate n samples from N(mu_true, sigma_true^2).
    2. Compute the sample mean and its two-sided confidence interval.
    3. Print results and plot histogram with CI bounds.
    """
    # 1. Generate data
    data = np.random.normal(loc=mu_true, scale=sigma_true, size=n)

    # 2. Sample statistics
    sample_mean = np.mean(data)
    sem = stats.sem(data)  # standard error of the mean

    # 3. Critical t-value
    df = n - 1
    alpha = 1 - confidence
    t_crit = stats.t.ppf(1 - alpha/2, df)

    # 4. Confidence interval
    margin = t_crit * sem
    ci_lower = sample_mean - margin
    ci_upper = sample_mean + margin

    # 5. Print results
    print(f"True mean           = {mu_true:.3f}")
    print(f"Sample mean         = {sample_mean:.3f}")
    print(f"{int(confidence*100)}% CI for mean = [{ci_lower:.3f}, {ci_upper:.3f}]")

    # 6. Plot
    plt.hist(data, bins=15, density=True, edgecolor='black', alpha=0.7, label='Sample data')
    x_vals = np.linspace(sample_mean - 4*sem, sample_mean + 4*sem, 200)
    plt.plot(x_vals, stats.norm.pdf(x_vals, mu_true, sigma_true), 'r-', label='True PDF')
    plt.axvline(sample_mean, linestyle='--', color='blue', label='Sample Mean')
    plt.axvline(ci_lower, linestyle=':', color='green', label='CI Lower')
    plt.axvline(ci_upper, linestyle=':', color='green', label='CI Upper')
    plt.title("Simulated Data & Confidence Interval")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()


if __name__ == "__main__":
    simulate_and_ci(mu_true=5.0, sigma_true=2.0, n=50, confidence=0.95)
````

````python
#24.  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
#distribution
import numpy as np

def compute_skewness_kurtosis(data):
    """
    Compute the sample skewness and excess kurtosis of a 1D dataset.

    Parameters
    ----------
    data : array-like
        Input data.

    Returns
    -------
    skewness : float
        The skewness of the data.
    kurtosis : float
        The excess kurtosis of the data (kurtosis minus 3).
    """
    arr = np.asarray(data)
    n = arr.size
    mean = arr.mean()
    std = arr.std(ddof=0)

    # Central moments
    m3 = np.sum((arr - mean)**3) / n
    m4 = np.sum((arr - mean)**4) / n

    # Skewness = m3 / std^3
    skewness = m3 / (std**3)

    # Excess kurtosis = m4 / std^4 - 3
    kurtosis = m4 / (std**4) - 3

    return skewness, kurtosis

if __name__ == "__main__":
    # Generate 1000 samples from a standard normal distribution
    data = np.random.normal(loc=0, scale=1, size=1000)

    skew, kurt = compute_skewness_kurtosis(data)
    print(f"Skewness: {skew:.4f}")
    print(f"Excess Kurtosis: {kurt:.4f}")
````

````python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
#interpret the results
import numpy as np
from scipy import stats

def z_test(sample, popmean, popstd, alpha=0.05, alternative='two-sided'):
    """
    Perform a one-sample Z-test comparing sample mean to a known population mean.

    Parameters
    ----------
    sample : array-like
        Your sample data.
    popmean : float
        The known population mean (Œº‚ÇÄ).
    popstd : float
        The known population standard deviation (œÉ).
    alpha : float, optional
        Significance level (default=0.05).
    alternative : {'two-sided', 'larger', 'smaller'}, optional
        The alternative hypothesis:
        - 'two-sided': mean ‚â† Œº‚ÇÄ
        - 'larger': mean > Œº‚ÇÄ
        - 'smaller': mean < Œº‚ÇÄ

    Returns
    -------
    z_stat : float
        The computed Z statistic.
    p_value : float
        The p-value for the test.
    reject : bool
        Whether to reject the null hypothesis at the given Œ±.
    """
    sample = np.asarray(sample)
    n = sample.size
    xbar = sample.mean()

    # Compute Z statistic
    z_stat = (xbar - popmean) / (popstd / np.sqrt(n))

    # Compute p-value
    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))
    elif alternative == 'larger':
        p_value = 1 - stats.norm.cdf(z_stat)
    elif alternative == 'smaller':
        p_value = stats.norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-sided', 'larger', or 'smaller'")

    reject = p_value < alpha
    return z_stat, p_value, reject

if __name__ == "__main__":
    # Example usage:
    # Generate a sample (e.g., from a population with true mean 100, sigma=15)
    np.random.seed(42)
    sample_data = np.random.normal(loc=102, scale=15, size=50)

    # Known population parameters
    mu0 = 100     # hypothesized population mean
    sigma = 15    # known population std dev

    # Perform the Z-test
    z, p, reject_null = z_test(
        sample=sample_data,
        popmean=mu0,
        popstd=sigma,
        alpha=0.05,
        alternative='two-sided'
    )

    # Display results
    print(f"Sample size        : {len(sample_data)}")
    print(f"Sample mean        : {sample_data.mean():.4f}")
    print(f"Population mean Œº‚ÇÄ : {mu0:.4f}")
    print(f"Population œÉ       : {sigma:.4f}")
    print(f"Z statistic        : {z:.4f}")
    print(f"p-value            : {p:.4f}")

    if reject_null:
        print(f"Result             : Reject H‚ÇÄ at Œ± = 0.05")
    else:
        print(f"Result             : Fail to reject H‚ÇÄ at Œ± = 0.05")
````

````python
#2.  Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python
import numpy as np
from scipy import stats

def simulate_and_test(n1=30, n2=30,
                      mu1=0.0, mu2=0.5,
                      sigma1=1.0, sigma2=1.0,
                      alpha=0.05, alternative='two-sided'):
    """
    Simulate two independent samples from normal distributions, perform a t-test,
    and return the t-statistic and p-value.

    Parameters
    ----------
    n1, n2 : int
        Sample sizes for group 1 and group 2.
    mu1, mu2 : float
        True means for group 1 and group 2.
    sigma1, sigma2 : float
        True standard deviations for group 1 and group 2.
    alpha : float
        Significance level for hypothesis test.
    alternative : {'two-sided', 'larger', 'smaller'}
        Type of alternative hypothesis.

    Returns
    -------
    t_stat : float
        The computed t statistic.
    p_value : float
        The p-value of the test.
    reject_null : bool
        True if we reject H‚ÇÄ at level Œ±; False otherwise.
    """
    # 1. Simulate random data
    sample1 = np.random.normal(loc=mu1, scale=sigma1, size=n1)
    sample2 = np.random.normal(loc=mu2, scale=sigma2, size=n2)

    # 2. Perform two-sample t-test (Welch‚Äôs t-test by default)
    t_stat, p_two = stats.ttest_ind(sample1, sample2, equal_var=False)

    # 3. Adjust p-value for one-sided tests if needed
    if alternative == 'two-sided':
        p_value = p_two
    elif alternative == 'larger':
        # H‚ÇÅ: mean1 > mean2
        # one-sided p = p_two / 2 if t_stat > 0 else 1 - (p_two / 2)
        p_value = stats.t.sf(t_stat,
                              df = ((sigma1**2/n1 + sigma2**2/n2)**2) /
                                   ((sigma1**4)/((n1**2)*(n1-1)) + (sigma2**4)/((n2**2)*(n2-1))))
    elif alternative == 'smaller':
        # H‚ÇÅ: mean1 < mean2
        p_value = stats.t.cdf(t_stat,
                              df = ((sigma1**2/n1 + sigma2**2/n2)**2) /
                                   ((sigma1**4)/((n1**2)*(n1-1)) + (sigma2**4)/((n2**2)*(n2-1))))
    else:
        raise ValueError("alternative must be 'two-sided', 'larger', or 'smaller'")

    # 4. Determine whether to reject H‚ÇÄ
    reject_null = p_value < alpha

    return t_stat, p_value, reject_null

if __name__ == "__main__":
    # Set random seed for reproducibility
    np.random.seed(0)

    # Simulate and test
    t_stat, p_val, reject = simulate_and_test(
        n1=30, n2=30,
        mu1=0.0, mu2=0.5,       # small true difference
        sigma1=1.0, sigma2=1.0,
        alpha=0.05,
        alternative='two-sided'
    )

    # Display results
    print(f"t statistic      : {t_stat:.4f}")
    print(f"p-value          : {p_val:.4f}")
    if reject:
        print(f"Decision         : reject H‚ÇÄ (significant at Œ± = 0.05)")
    else:
        print(f"Decision         : fail to reject H‚ÇÄ (not significant at Œ± = 0.05)")

````

````python
#3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean?
import numpy as np
from scipy import stats

def one_sample_z_test(sample, popmean, popstd, alpha=0.05, alternative='two-sided'):
    """
    Perform a one-sample Z-test comparing the sample mean to a known population mean.

    Parameters
    ----------
    sample : array-like
        The sample observations.
    popmean : float
        The known or hypothesized population mean (Œº‚ÇÄ).
    popstd : float
        The known population standard deviation (œÉ).
    alpha : float, optional
        Significance level (default is 0.05).
    alternative : {'two-sided', 'greater', 'less'}, optional
        Defines the alternative hypothesis:
        - 'two-sided': mean ‚â† Œº‚ÇÄ
        - 'greater'  : mean > Œº‚ÇÄ
        - 'less'     : mean < Œº‚ÇÄ

    Returns
    -------
    z_stat : float
        The computed Z statistic.
    p_value : float
        The p-value of the test.
    reject_null : bool
        True if you reject H‚ÇÄ at the given Œ±; False otherwise.
    """
    sample = np.asarray(sample)
    n = sample.size
    xbar = sample.mean()

    # Compute Z statistic
    z_stat = (xbar - popmean) / (popstd / np.sqrt(n))

    # Compute p-value
    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))
    elif alternative == 'greater':
        p_value = 1 - stats.norm.cdf(z_stat)
    elif alternative == 'less':
        p_value = stats.norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-sided', 'greater', or 'less'")

    reject_null = (p_value < alpha)
    return z_stat, p_value, reject_null

if __name__ == "__main__":
    # Example: suppose population mean Œº‚ÇÄ = 100, œÉ = 12
    np.random.seed(1)
    sample_data = np.random.normal(loc=102, scale=12, size=40)

    z, p, reject = one_sample_z_test(
        sample=sample_data,
        popmean=100,
        popstd=12,
        alpha=0.05,
        alternative='two-sided'
    )

    print(f"Sample size      : {len(sample_data)}")
    print(f"Sample mean      : {sample_data.mean():.3f}")
    print(f"Population mean  : {100:.3f}")
    print(f"Population œÉ     : {12:.3f}")
    print(f"Z statistic      : {z:.3f}")
    print(f"P-value          : {p:.4f}")
    if reject:
        print("Result           : Reject H‚ÇÄ (significant difference)")
    else:
        print("Result           : Fail to reject H‚ÇÄ (no significant difference)")
````

````python
#4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def two_tailed_z_test_with_plot(sample, popmean, popstd, alpha=0.05):
    """
    Performs a two-tailed Z-test and visualizes the decision regions.

    Parameters
    ----------
    sample : array-like
        Sample data.
    popmean : float
        Known population mean.
    popstd : float
        Known population standard deviation.
    alpha : float
        Significance level (default 0.05).

    Returns
    -------
    z_stat : float
        Computed Z statistic.
    p_value : float
        Two-tailed p-value.
    """
    sample = np.asarray(sample)
    n = sample.size
    xbar = sample.mean()

    # Compute Z statistic
    z_stat = (xbar - popmean) / (popstd / np.sqrt(n))
    # Two-tailed p-value
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    # Critical value
    z_crit = stats.norm.ppf(1 - alpha/2)

    # Plot standard normal PDF
    x = np.linspace(-4, 4, 400)
    y = stats.norm.pdf(x, 0, 1)

    plt.figure()
    plt.plot(x, y, label="Standard Normal PDF")
    plt.fill_between(x, y, where=(x <= -z_crit) | (x >= z_crit), color='orange', alpha=0.3,
                     label=f"Rejection regions (Œ±={alpha})")
    plt.axvline(-z_crit, color='orange', linestyle='-', label=f"Critical z = ¬±{z_crit:.2f}")
    plt.axvline( z_crit, color='orange', linestyle='-')
    plt.axvline(z_stat, color='black', linestyle='--', label=f"Observed z = {z_stat:.2f}")

    plt.title("Two-Tailed Z-Test Decision Regions")
    plt.xlabel("Z value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.tight_layout()
    plt.show()

    return z_stat, p_value

if __name__ == "__main__":
    # Example sample drawn from N(5.5, 1.2)
    np.random.seed(42)
    sample_data = np.random.normal(loc=5.5, scale=1.2, size=40)

    z_stat, p_val = two_tailed_z_test_with_plot(
        sample=sample_data,
        popmean=5.0,
        popstd=1.2,
        alpha=0.05
    )

    print(f"Z statistic: {z_stat:.4f}")
    print(f"Two-tailed p-value: {p_val:.4f}")
````

````python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def plot_type_errors(mu0=0, mu1=1, sigma=1, n=30, alpha=0.05):
    """
    Calculate and visualize Type I (Œ±) and Type II (Œ≤) errors for a one-sided test.

    Parameters
    ----------
    mu0 : float
        Mean under the null hypothesis H‚ÇÄ.
    mu1 : float
        Mean under the alternative hypothesis H‚ÇÅ.
    sigma : float
        Known population standard deviation.
    n : int
        Sample size.
    alpha : float
        Significance level (Type I error rate).

    Returns
    -------
    alpha : float
        Type I error rate.
    beta : float
        Type II error rate.
    """
    # Standard error of the mean
    se = sigma / np.sqrt(n)

    # Critical z-value and the corresponding sample‚Äêmean threshold
    z_crit = stats.norm.ppf(1 - alpha)
    x_crit = mu0 + z_crit * se

    # Compute Type II error (Œ≤)
    beta = stats.norm.cdf(x_crit, loc=mu1, scale=se)

    # Prepare x‚Äêaxis for plotting
    x_min = mu0 - 4 * se
    x_max = mu1 + 4 * se
    x = np.linspace(x_min, x_max, 500)

    # PDFs under H‚ÇÄ and H‚ÇÅ
    y_null = stats.norm.pdf(x, loc=mu0, scale=se)
    y_alt  = stats.norm.pdf(x, loc=mu1, scale=se)

    # Plot distributions
    plt.figure()
    plt.plot(x, y_null, label="Null (H‚ÇÄ)", linewidth=2)
    plt.plot(x, y_alt,  label="Alternative (H‚ÇÅ)", linestyle='--', linewidth=2)

    # Shade Type I error region (Œ±) under H‚ÇÄ
    plt.fill_between(x, y_null, where=(x >= x_crit), alpha=0.3,
                     label=f"Type I error Œ± = {alpha:.2f}")
    # Shade Type II error region (Œ≤) under H‚ÇÅ
    plt.fill_between(x, y_alt, where=(x <= x_crit), alpha=0.3,
                     label=f"Type II error Œ≤ = {beta:.2f}")

    # Mark critical threshold
    plt.axvline(x_crit, linestyle=':', color='black', label=f"Critical value = {x_crit:.2f}")

    plt.title("Type I and Type II Error Regions")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.legend()
    plt.tight_layout()
    plt.show()

    return alpha, beta

if __name__ == "__main__":
    # Example: H‚ÇÄ mean = 0, H‚ÇÅ mean = 1, œÉ = 1, n = 30, Œ± = 0.05
    alpha, beta = plot_type_errors(mu0=0, mu1=1, sigma=1, n=30, alpha=0.05)
    print(f"Type I error (Œ±): {alpha:.3f}")
    print(f"Type II error (Œ≤): {beta:.3f}")
````

````python
#6. Write a Python program to perform an independent T-test and interpret the results?
import numpy as np
from scipy import stats

def independent_t_test(sample1, sample2, alpha=0.05, alternative='two-sided'):
    """
    Perform an independent two-sample t-test (Welch‚Äôs t-test by default).

    Parameters
    ----------
    sample1 : array-like
        Observations from group 1.
    sample2 : array-like
        Observations from group 2.
    alpha : float, optional
        Significance level (default = 0.05).
    alternative : {'two-sided', 'greater', 'less'}, optional
        Defines the alternative hypothesis:
        - 'two-sided': Œº1 ‚â† Œº2
        - 'greater'  : Œº1 > Œº2
        - 'less'     : Œº1 < Œº2

    Returns
    -------
    t_stat : float
        The t statistic.
    p_value : float
        The p-value of the test (adjusted for one- or two-sided).
    reject_null : bool
        True if we reject H‚ÇÄ at the given Œ±.
    """
    # Calculate Welch‚Äôs t-test
    t_stat, p_two = stats.ttest_ind(sample1, sample2, equal_var=False)

    # Adjust p-value for one-sided tests
    if alternative == 'two-sided':
        p_value = p_two
    elif alternative == 'greater':
        # H‚ÇÅ: Œº1 > Œº2
        p_value = p_two / 2 if t_stat > 0 else 1 - p_two / 2
    elif alternative == 'less':
        # H‚ÇÅ: Œº1 < Œº2
        p_value = p_two / 2 if t_stat < 0 else 1 - p_two / 2
    else:
        raise ValueError("alternative must be 'two-sided', 'greater', or 'less'")

    reject_null = p_value < alpha
    return t_stat, p_value, reject_null

if __name__ == "__main__":
    # Example data: two independent groups
    np.random.seed(0)
    group1 = np.random.normal(loc=5.0, scale=1.5, size=30)
    group2 = np.random.normal(loc=4.5, scale=1.5, size=30)

    # Perform test
    t_stat, p_val, reject = independent_t_test(
        group1, group2,
        alpha=0.05,
        alternative='two-sided'
    )

    # Output results
    print(f"Group 1 mean = {group1.mean():.3f}, n = {len(group1)}")
    print(f"Group 2 mean = {group2.mean():.3f}, n = {len(group2)}")
    print(f"t-statistic   = {t_stat:.3f}")
    print(f"p-value       = {p_val:.4f}")

    if reject:
        print(f"Result        : Reject H‚ÇÄ at Œ± = 0.05 ‚Üí means differ significantly.")
    else:
        print(f"Result        : Fail to reject H‚ÇÄ at Œ± = 0.05 ‚Üí no significant difference.")
````

````python
#8. Simulate data and perform both Z-test and T-test, then compare the results using Python
import numpy as np
from scipy import stats

def one_sample_z_test(sample, popmean, popstd, alpha=0.05):
    """
    Perform a one-sample Z-test.
    Returns (z_stat, p_value, reject_null).
    """
    n = len(sample)
    xbar = sample.mean()
    z_stat = (xbar - popmean) / (popstd / np.sqrt(n))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))
    return z_stat, p_value, (p_value < alpha)

def one_sample_t_test(sample, popmean, alpha=0.05):
    """
    Perform a one-sample T-test.
    Returns (t_stat, p_value, reject_null).
    """
    t_stat, p_value = stats.ttest_1samp(sample, popmean)
    return t_stat, p_value, (p_value < alpha)

if __name__ == "__main__":
    # Simulation parameters
    np.random.seed(42)
    pop_mean = 50
    pop_std  = 5      # assume known for Z-test
    n_samples = 30
    alpha = 0.05

    # 1. Simulate data from N(pop_mean, pop_std)
    sample = np.random.normal(loc=pop_mean, scale=pop_std, size=n_samples)

    # 2. Perform Z-test (using known pop_std)
    z_stat, z_p, z_reject = one_sample_z_test(sample, pop_mean, pop_std, alpha)

    # 3. Perform T-test (pop_std unknown)
    t_stat, t_p, t_reject = one_sample_t_test(sample, pop_mean, alpha)

    # 4. Print and compare results
    print(f"Sample size         : {n_samples}")
    print(f"Sample mean         : {sample.mean():.4f}")
    print(f"Population mean     : {pop_mean}")
    print()
    print("=== Z-test (œÉ known) ===")
    print(f"Z statistic         : {z_stat:.4f}")
    print(f"p-value             : {z_p:.4f}")
    print(f"Reject H‚ÇÄ?          : {'Yes' if z_reject else 'No'} (Œ± = {alpha})")
    print()
    print("=== T-test (œÉ estimated) ===")
    print(f"T statistic         : {t_stat:.4f}")
    print(f"p-value             : {t_p:.4f}")
    print(f"Reject H‚ÇÄ?          : {'Yes' if t_reject else 'No'} (Œ± = {alpha})")
````

````python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
from scipy import stats

def confidence_interval_mean(data, confidence=0.95):
    """
    Calculate the two-sided confidence interval for the sample mean.

    Parameters
    ----------
    data : array-like
        The sample observations.
    confidence : float, optional
        The confidence level (default is 0.95 for a 95% CI).

    Returns
    -------
    (mean, lower_bound, upper_bound) : tuple of floats
        mean           : the sample mean
        lower_bound    : lower bound of the confidence interval
        upper_bound    : upper bound of the confidence interval
    """
    arr = np.asarray(data)
    n = arr.size
    mean = arr.mean()
    sem = stats.sem(arr)  # standard error of the mean (uses ddof=1)

    # t critical value for two-sided interval
    alpha = 1 - confidence
    t_crit = stats.t.ppf(1 - alpha/2, df=n-1)

    margin = t_crit * sem
    lower = mean - margin
    upper = mean + margin

    return mean, lower, upper

if __name__ == "__main__":
    # Example usage
    np.random.seed(0)
    sample = np.random.normal(loc=100, scale=15, size=50)

    mean, lo, hi = confidence_interval_mean(sample, confidence=0.95)
    print(f"Sample size: {len(sample)}")
    print(f"Sample mean: {mean:.2f}")
    print(f"95% CI for the mean: ({lo:.2f}, {hi:.2f})")
````

````python
#10.  Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
from scipy import stats

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for the sample mean at a given confidence level.

    Parameters
    ----------
    data : array-like
        The sample observations.
    confidence : float, optional
        The desired confidence level (default=0.95).

    Returns
    -------
    moe : float
        The margin of error.
    """
    arr = np.asarray(data)
    n = arr.size
    sem = stats.sem(arr)  # standard error of the mean (ddof=1)

    alpha = 1 - confidence
    # Use t-distribution critical value for small samples
    t_crit = stats.t.ppf(1 - alpha/2, df=n-1)

    moe = t_crit * sem
    return moe

if __name__ == "__main__":
    # Example usage
    np.random.seed(1)
    sample = np.random.normal(loc=20, scale=5, size=40)

    confidence_level = 0.95
    moe = margin_of_error(sample, confidence=confidence_level)
    sample_mean = sample.mean()

    print(f"Sample size           : {len(sample)}")
    print(f"Sample mean           : {sample_mean:.3f}")
    print(f"{int(confidence_level*100)}% margin of error : ¬±{moe:.3f}")
    print(f"{int(confidence_level*100)}% confidence interval: ({sample_mean - moe:.3f}, {sample_mean + moe:.3f})")
````

#11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
```python
import numpy as np

def bayes_update(prior, likelihood):
    """
    Perform a single Bayes‚Äô Theorem update.

    Parameters
    ----------
    prior : dict
        Prior probabilities for each hypothesis, e.g. {'H1': 0.5, 'H2': 0.5}.
        Values must sum to 1.
    likelihood : dict
        Likelihood of the observed data under each hypothesis,
        e.g. {'H1': P(data|H1), 'H2': P(data|H2)}.

    Returns
    -------
    posterior : dict
        Posterior probabilities for each hypothesis after observing the data.
    """
    # Compute unnormalized posteriors
    unnorm = {h: prior[h] * likelihood[h] for h in prior}
    # Evidence is sum over unnormalized posteriors
    evidence = sum(unnorm.values())
    # Normalize to get proper probabilities
    posterior = {h: unnorm[h] / evidence for h in unnorm}
    return posterior

if __name__ == "__main__":
    # Example: A medical test for a rare disease
    # Hypotheses: D = "has disease", ¬¨D = "does not have disease"
    prior = {
        'D' : 0.01,   # 1% of population has the disease
        '¬¨D': 0.99
    }
    # Likelihoods: P(test positive | hypothesis)
    likelihood = {
        'D' : 0.95,   # true positive rate
        '¬¨D': 0.05    # false positive rate
    }

    posterior = bayes_update(prior, likelihood)
    print("Posterior probabilities after a positive test:")
    for h, p in posterior.items():
        print(f"  P({h} | test+) = {p:.4f}")

    # If you observe a second independent positive test:
    likelihood2 = {
        'D' : 0.95,
        '¬¨D': 0.05
    }
    # Use previous posterior as new prior
    posterior2 = bayes_update(posterior, likelihood2)
    print("\nAfter two positive tests:")
    for h, p in posterior2.items():
        print(f"  P({h} | two tests+) = {p:.4f}")
```

---

## Explanation of the Process

1. **Define Hypotheses and Priors**

   * List all mutually exclusive hypotheses (e.g., having the disease `D` or not `¬¨D`).
   * Assign a **prior probability** to each, reflecting your belief *before* seeing the data.

2. **Specify Likelihoods**

   * For each hypothesis, compute or estimate the probability of observing the new data under that hypothesis:

     $$
       P(\text{data} \mid H_i).
     $$

3. **Apply Bayes‚Äô Theorem**

   * For each hypothesis $H_i$, compute the *unnormalized* posterior:

     $$
       \text{unnorm}_i = P(H_i) \times P(\text{data} \mid H_i).
     $$
   * Compute the **evidence** (total probability of the data):

     $$
       P(\text{data}) = \sum_j \text{unnorm}_j.
     $$
   * Normalize to get the **posterior** probabilities:

     $$
       P(H_i \mid \text{data})
       = \frac{\text{unnorm}_i}{P(\text{data})}.
     $$

4. **Sequential Updates**

   * If you get more data later (e.g., a second test result), feed the previous posterior back in as the new prior and repeat the process

````python
#13.  Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

def compute_expected_frequencies(observed):
    """
    Compute the expected frequencies for a Chi-square test of independence.

    Parameters
    ----------
    observed : array-like or pandas.DataFrame
        The observed contingency table. If a 2D array or list of lists is passed,
        it will be converted to a DataFrame with integer row/column labels.

    Returns
    -------
    expected_df : pandas.DataFrame
        The table of expected frequencies under the null hypothesis of independence.
    """
    # If input is not a DataFrame, convert it
    if not isinstance(observed, pd.DataFrame):
        observed = pd.DataFrame(observed)

    # Use scipy to get expected counts
    _, _, _, expected = chi2_contingency(observed, correction=False)

    # Wrap expected array back into a DataFrame with the same index/columns
    expected_df = pd.DataFrame(expected,
                               index=observed.index,
                               columns=observed.columns)
    return expected_df

if __name__ == "__main__":
    # Example observed data: survey counts by Gender and Preference
    data = {
        'Product A': [30, 20],  # Male, Female
        'Product B': [20, 25],
        'Product C': [10, 15]
    }
    observed_df = pd.DataFrame(data, index=['Male', 'Female'])

    print("Observed Frequencies:")
    print(observed_df, "\n")

    expected_df = compute_expected_frequencies(observed_df)

    print("Expected Frequencies (under independence):")
    print(expected_df)
````

````python
#14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
from scipy.stats import chisquare

def goodness_of_fit_test(observed, expected=None):
    """
    Perform a chi-square goodness-of-fit test.

    Parameters
    ----------
    observed : list or array-like
        Observed frequencies for each category.
    expected : list or array-like, optional
        Expected frequencies. If None, assumes a uniform distribution.

    Returns
    -------
    chi2_stat : float
        Chi-square test statistic.
    p_value : float
        P-value for the test.
    """
    chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)
    return chi2_stat, p_value

if __name__ == "__main__":
    # Example: Dice roll outcomes (observed vs expected)
    observed_counts = [10, 8, 12, 9, 11, 10]  # counts for 6 faces of a die
    expected_counts = [10] * 6  # expected uniform distribution

    chi2, p = goodness_of_fit_test(observed_counts, expected_counts)

    print("Observed Counts:", observed_counts)
    print("Expected Counts:", expected_counts)
    print(f"\nChi-square Statistic: {chi2:.4f}")
    print(f"P-value            : {p:.4f}")

    alpha = 0.05
    if p < alpha:
        print("\nConclusion: Reject the null hypothesis ‚Äî the data does NOT follow the expected distribution.")
    else:
        print("\nConclusion: Fail to reject the null hypothesis ‚Äî the data fits the expected distribution.")
````
